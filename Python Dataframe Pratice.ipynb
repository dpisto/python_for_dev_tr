## Creating a basic dataframe from list

```python
# define the data as a list

data = [
    ("Dexter","Johsons","dog","shiba inu","red sesame",1.5,35,"m",False,"both",True),
    ("Alfred","Johsons","cat","mix","tuxedo",4,12,"m",False,"indoor",True),
    ("Petra","Smith","cat","ragdoll","calico",6,None,"f",False,"both",True),
    ("Ava","Smith","dog","mix","blk/wht",12,32,"f",True,"both",True),
    ("Schroder","Brown","cat","mix","orange",13,15,"m",False,"indoor",True),
    ("Blackbeard","Brown","bird","rparrot","multi",5,3,"f",False,"indoor",),
]

# define labels

labels = ["name","owner","type","breed","color","age","weight","gender","health issues","indoor/outdoor","vaccinated"]

#create dataframe

vet_records = pd.DataFrame.from_records(data, columns=labels)
```

### Examining the data in a dataframe

There are several different ways to examine data using a pandas dataframe. Two are `.head()` and `.tail()`. These show the first five and the last five rows of the dataframe respectively.

```python
# displays the first five rows in the data frame

vet_records.head()    
```

```python
# diplays the last five rows in the data frame

vet_records.tail()
```

```python
# displays all the records of the data frame

vet_records
```

**`.dtypes` show you the types of data in the dataframe by column. If the `dtype` is `object`, this indicates that pandas is seeing that data as more than one type.**

```python
vet_records.dtypes
```

**`.describe` shows statistical operations on columns that these operations can be performed on**

```python
vet_records.describe()
```

```python
vet_records.describe(include="all")
```

**`.at` allows the usewr to change the value of a specific cell**

```python
vet_records.at[0, "weight"] = 34.7
vet_records
```

**`.assign` is used to add another column of data**

```python
vet_records = vet_records.assign(age_weight=(vet_records['age']/vet_records['weight']))
```

```python
vet_records
```
